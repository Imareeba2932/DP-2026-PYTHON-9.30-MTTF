BREAK & CONTINUE

```python
fruits = ['apple','banana','grapes','onion','mango','orange']
for x in fruits:
    if x == 'onion':
        break
    print(x)
```

```python
fruits = ['apple','banana','grapes','onion','mango','orange']
for x in fruits:
    if x == 'onion':
        continue
    print(x)
```
