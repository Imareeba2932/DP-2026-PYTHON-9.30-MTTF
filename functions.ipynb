`TYPES OF FUNCTIONS`
- user defined functions
    - parameterized function
    - non parameterized function
    - return function
    - non return function
    - default parameterized function
    - lambda function
- built in function

```python
#non parameterized function
def greet(): #function initialiaze
    print("Hello, World!")
    print("Welcome to Python programming.")
    print("Have a great day!")
    print("Goodbye!")

greet() #function call
```

```python
a = 10
b = 20
print(a+b)
```

```python
#paramterized function
def sum(x,y): #parameters
    print(x+y)

sum(10,20) #arguments
```

```python
#non return function
def sum(x,y): #parameters
    print(x+y)

sum = sum(10,20)
print(sum)
```

```python
#return function
def sum(x,y): #parameters
    return x+y

sum = sum(10,20)
print(sum)
```

```python
print(sum)
```

```python
def avg():
    print(sum/2)

avg()
```

```python
def my_country(country="India"):
    print(f"I love {country}")

my_country("USA")
my_country("UK")
my_country()
```
