- how to access key value pair 
- how to add or update a pair
- loops in dictionary
- nested dictionary

```python
student = {
    "name" : "Alice",
    "age" : input("Enter age: "),
    "city" : "Lucknow"
}
print(student)
```

```python
info = dict(name = "Alice", age = input("Enter age: "), city = "Lucknow")
print(info)
```

```python
temp = {
    'lucknow' : [28, 30, 25],
    'delhi' : [35, 38, 40],
    'chennai' : [32, 34, 31]
}
print(temp)
```

```python
#1st method to access value
info["name"]
temp['lucknow']
```

```python
#2nd method to access value
info.get("name")
```

```python
#add new key-value pair
info["country"] = "India"
print(info)
```

```python
#update value of existing key
info["age"] = 20
print(info)
```

```python
#delete key-value pair
del info["country"]
print(info)
```

```python
print(info.keys())
print(info.values())
print(info.items())
```

```python
for k,v in info.items():
    print(f'{k} : {v}')
```
