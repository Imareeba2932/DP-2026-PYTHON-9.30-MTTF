LIST METHODS

- append() - to add item at the end of the list
- insert() - to add item at any specific position in the list
- pop() - to delete item using its index position
- remove() - to delete items using item name
- copy() - to make a duplicate of the list
- sort() - to sort the list in a particular order
- extend() - to combine two list into a single list
- clear() - to delete all the items from the list but not the list

```python
apps.append('amazon')
print(apps)
```

```python
apps.insert(2,'flipkart')
print(apps)
```

```python
apps.pop()
print(apps)
```

```python
apps.pop(5)
print(apps)
```

```python
apps.remove('watsap')
print(apps)
```

```python
apps.insert(5, 'Netflix')
```

```python
apps.insert(7, 'Hotstar')
```

```python
print(apps)
```

```python
apps.sort()
print(apps)
```

```python
apps2 = apps.copy()
print(apps2)
print(apps)
```

```python
apps2.clear()
print(apps2)
```

```python
a = [1,2,3,4,5]
apps.extend(a)
print(apps)
```

```python
a.extend(apps)
print(a)
```
