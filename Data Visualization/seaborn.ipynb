```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
#histogram - age distribution
sns.histplot(df, x='age',color='red',kde=True) #kernel density estimate
plt.title("Age distribution")
plt.show()
```

```python
#countplot of passengers based on gender
sns.countplot(df, x='sex', palette = 'coolwarm')
plt.title("Count of passengers based on gender")
```

```python
#histogram - age distribution
import numpy as np

sns.histplot(df, x='age',color='red',kde=True) #kernel density estimate
plt.xticks(np.arange(0,df['age'].max()+1, 5))
plt.title("Age distribution")
plt.show()
```

```python
#count of passengers based on passenger class
#bar chart of age vs fare
```

```python
plt.figure(figsize=(15,8))
sns.barplot(df[df['age']>40],x='age', y='fare',palette='coolwarm')
```
