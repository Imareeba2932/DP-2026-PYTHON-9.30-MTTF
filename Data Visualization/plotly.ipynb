```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
df
```

pip install nbformat

```python
#total accidents statewise
fig = px.bar(df, x='abbrev', y= 'total',
            title = 'Total Accidents Statewise',
            labels = {'abbrev' : 'State','total' : 'Total Accidents'},
            template = 'plotly_dark',
            color = 'abbrev'
            )
fig.show()
```

```python
#top10 states by insurance premium
#top10 = df.nlargest(10,'ins_premium')
top10 = df.sort_values('ins_premium').tail(10)
fig = px.bar(top10, x='abbrev',y='ins_premium',
            title = 'Top10 states by insurance presmium',
            template = 'plotly_dark')
fig.show()
```

```python
#percentage of speeding accidents
fig = px.pie(df, values='speeding', names='abbrev',
            title = 'Percentage of speeding accidents',
            color_discrete_sequence = px.colors.sequential.Purples_r)
fig.update_traces(textposition = 'inside')
fig.show()
```

```python
#Alcohol vs Speeding
fig = px.scatter(df, x='alcohol', y='speeding',
                color = 'abbrev',
                title = 'Alcohol vs Speeding',
                size = 'total',template = 'plotly_dark')
fig.update_layout(title={'x' : 0.5})
fig.show()
```

```python
df = sns.load_dataset('tips')
df
```

```python
import matplotlib.pyplot as plt
```

```python
#pairplot
sns.pairplot(df)
plt.show()
```

```python
#heatmap
corr = df.corr(numeric_only = True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.columns
```

```python
avg_bill = df.groupby('day')["total_bill"].mean().reset_index()
fig = px.line(avg_bill, x ="day",y="total_bill")
fig.show()

```

```python
fig = px.sunburst(df, path=['day','time','smoker'],
                  title = 'Sunburst Chart of Tips Dataset',)
fig.show()
```
