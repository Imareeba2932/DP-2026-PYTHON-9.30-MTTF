```python
import matplotlib.pyplot as plt
```

```python
#creating a figure and axes
fig, ax = plt.subplots(figsize=(4,2))
#assigning background color
fig.set_facecolor("yellow")
ax.set_facecolor("lightgreen")
#plotting data
x = [1,2,3,4,5,6,7,8,9]
y = [4,7,2,5,10,14,6,3,11]
plt.plot(x,y,
         color="red",
         linewidth=1,
         linestyle = "--",
         marker='o',
         markersize = 5,
         markerfacecolor = 'white',
         markeredgecolor= 'black',
         markeredgewidth = 2
         )

ax.set_title("Custom Graph", color="red", fontsize = 15)
ax.set_xlabel("X axis", color="red", fontsize = 10)
ax.set_ylabel("Y axis", color="red", fontsize = 10)

#adding text
ax.text(5,6, 'text', color='purple',fontsize='10')

#adding vertical and horizontal lines
ax.axvline(x=5, color='black', linestyle='--')
ax.axhline(y=8, color='black', linestyle='-.')
```
