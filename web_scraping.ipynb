```python
from dputils.scrape import Scraper, Tag
```

```python
url = "https://www.flipkart.com/search?q=cosmic%20byte%20controller&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
scraper = Scraper(url)
```

```python
target_tag = Tag(name='div', cls='_75nlfW')
item_tag = Tag(name='div', cls='slAVV4')
title_tag = Tag(name='a', cls='wjcEIp', output='text')
price_tag = Tag(name='div', cls='Nx9bqj', output='text')
link_tag = Tag(name='a', cls='wjcEIp', output='href')
img_tag = Tag(name='img', cls='DByuf4', output='src')
```

```python
products = scraper.get_repeating_data_from_page(
    target=target_tag,
    items=item_tag,
    title=title_tag,
    price=price_tag,
    link=link_tag,
    img=img_tag
)
for product in products:
    print(product)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(products)
df.head()
```

```python
df.to_csv('controller_data.csv', index=False)
```
